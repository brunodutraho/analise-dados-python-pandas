```python
# 1-Análise de Desempenho por Time
desempenho_info = df[['Times', 'Pontos', 'Vitorias', 'Saldo_Gols', 'Gols_Pro', 'Gols_Contra']].set_index('Times')
desempenho_info
```

```python
# 2-Time com mais vitória
time_mais_vitoria = df[df['Vitorias'] == df['Vitorias'].max()]
time_mais_vitoria
```

```python
# 3-Time com menos vitoria
time_menos_vitoria = df[df['Vitorias'] == df['Vitorias'].min()]
time_menos_vitoria
```

```python
# 4-Distribuição de GOLs Pró
df[['Times', 'Gols_Pro']].sort_values(by='Gols_Pro', ascending=False)
```

```python
# 5-Relação entre gols pró e pontos
correlacao = df['Gols_Pro'].corr(df['Pontos'])
correlacao
```
