```python
import pandas as pd
```

```python
dados_vendas = {
    'Produto': ['Camisa', 'Calça', 'Saia', 'Camisa', 'Calça', 'Saia', 'Camiseta', 'Calça', 'Saia'],
    'Preço': [50, 100, 80, 55, 110, 85, 45, 95, 90],
    'Quantidade': [10, 15, 20, 5, 10, 15, 30, 20, 10],
    'Mês': ['Jan', 'Jan', 'Jan', 'Fev', 'Fev', 'Fev', 'Mar', 'Mar', 'Mar']
}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

```python
# Calculando a receita total por produto
df_vendas['Receita'] = df_vendas['Preço'] * df_vendas['Quantidade']
df_vendas
```

```python
# Sumarizando a receita total por produto
resumo_vendas = df_vendas.groupby('Produto')['Receita'].sum().reset_index()
resumo_vendas
```

```python
# Filtrar produtos com receita acima de R$ 1.000,00
produto_altas_vendas = resumo_vendas[resumo_vendas['Receita'] > 1000]
produto_altas_vendas
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.bar(resumo_vendas['Produto'], resumo_vendas['Receita'], color='skyblue')
plt.title('Receita Total por Produto')
plt.xlabel('Produto')
plt.ylabel('Receita R$')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight de Negócio

O produto **Calça** apresenta a maior receita total no período analisado,
seguido por **Saia**, indicando que esses itens concentram o maior volume
financeiro das vendas.
