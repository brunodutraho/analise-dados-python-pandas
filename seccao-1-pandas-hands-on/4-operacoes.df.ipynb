```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Idade': [25, 30, 35, 40, 27],
    'Cargo': ['Analista', 'Gerente', 'CEO', 'Analista', 'Coordenador'],
    'Salário': [5000, 8000, 15000, 4800, 6000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df[df['Salário'] > 5000 ]
```

```python
# Ordenam de forma decrescente se False e crescente se True
df.sort_values(by='Idade', ascending=False)
```

```python
df['Bônus'] = df['Salário'] * 0.1
df
```

```python
df.groupby('Cargo')['Salário'].mean()
```

### Insight de Negócio

O cargo de **CEO** apresenta a maior média salarial, seguido pelo cargo de
**Gerente**, evidenciando uma hierarquia clara de remuneração na empresa.
Cargos operacionais como **Analista** possuem menor média salarial.


```python
df.plot(
    kind='bar',
    x='Nome',
    y='Salário',
    title='Salário dos Funcionários',
    rot=45,
    color='green'
)
```

### Insight Visual

O gráfico de barras evidencia a diferença salarial entre os funcionários,
destacando **Charlie** como o maior salário individual do conjunto de dados.
